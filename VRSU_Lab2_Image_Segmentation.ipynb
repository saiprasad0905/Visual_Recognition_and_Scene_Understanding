```python
'''
Calculation thresold value is by otsu algorithm
thresold value decides the color
cv2.THRESH_OTSU calculate thresold using otsu algorithm
cv is computer vision library
'''
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("/content/Avg_filtering.jpeg")
if img is None:
  raise ValueError("Image not found")

# Convert the image to grayscale
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

thresold_value, thresold_image = cv2.threshold(grayscale_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print("Optimal thresold value:",thresold_value)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Histogram")
plt.hist(grayscale_img.ravel(), 256) # Plot histogram of grayscale image
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(1, 3, 3)
plt.title("Otsu Binarized Image")
plt.imshow(thresold_image, cmap='gray') # Use thresold_image instead of binary_img
plt.axis("off")

plt.show()
```
