```python
#Point processing(Lab1)
#Histogram Equalisation
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("/content/blurring3.jpg",0)#reading the image
he_img = cv2.equalizeHist(img)
plt.figure(figsize=(12,5))#size of histogram i.e. x and y length
plt.subplot(1,3,1)#dimensions of the original image
plt.title("Original")
plt.imshow(img,cmap = 'gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Clarified image")
plt.imshow(he_img,cmap ='gray')
plt.axis("off")#axis point numbers won't be visible

plt.subplot(1,3,3)
plt.title("Graph")#in the graph the y-axis indicates the white color pixels and x-axis indicates the gray color
plt.hist(he_img.ravel(),256)
plt.show()
```

```python
#point processing
import cv2
import matplotlib.pyplot as plt
#load grayscale image
img = cv2.imread("/content/clahe.jpeg",0)
#create CLAHE object
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
#apply enhancement
clahe_img = clahe.apply(img)
#plot
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img,cmap = 'gray')
plt.axis("off")
plt.subplot(1,3,2)
plt.title("CLAHE image")
plt.imshow(clahe_img,cmap ='Reds_r')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Histogram")
plt.hist(clahe_img.ravel(),256)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img= cv2.imread("/content/clahe2.jpeg") # Corrected filename
img_ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

y,cr,cb=cv2.split(img_ycrcb)

clahe=cv2.createCLAHE(clipLimit=2.5,tileGridSize=(8,8))
y_clahe=clahe.apply(y)

img_clahe=cv2.merge((y_clahe,cr,cb))

img_clahe=cv2.cvtColor(img_clahe,cv2.COLOR_YCrCb2BGR)

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("CLAHE Image")
plt.imshow(cv2.cvtColor(img_clahe, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Histogram of CLAHE Y-channel")
plt.hist(y_clahe.ravel(),256)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img_avg= cv2.imread("/content/Avg_filtering.jpeg")
img_gaussian= cv2.imread("/content/Gaussian_filtering.png")
img_median= cv2.imread("/content/median_filtering.jpeg")

if img_avg is None or img_gaussian is None or img_median is None:
    print("Error loading images.")
    exit()


img_avg_rgb=cv2.cvtColor(img_avg,cv2.COLOR_BGR2RGB)
img_gaussian_rgb=cv2.cvtColor(img_gaussian,cv2.COLOR_BGR2RGB)
img_median_rgb=cv2.cvtColor(img_median,cv2.COLOR_BGR2RGB)



#Apply Filters

#Average
avg_blur=cv2.blur(img_avg,(5,5))
avg_blur_rgb=cv2.cvtColor(avg_blur,cv2.COLOR_BGR2RGB)

#Guassian
gaussian_blur=cv2.GaussianBlur(img_gaussian,(15,15),0)
gaussian_blur_rgb=cv2.cvtColor(gaussian_blur,cv2.COLOR_BGR2RGB)

#Median
median_blur=cv2.medianBlur(img_median,5)
median_blur_rgb=cv2.cvtColor(median_blur,cv2.COLOR_BGR2RGB)


#Display Results

plt.figure(figsize=(14,10))


plt.figure(figsize=(14, 10))

# ----- Average Blur -----
plt.subplot(3, 2, 1)
plt.imshow(img_avg_rgb)
plt.title("Original Image (Average Blur)")
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(avg_blur_rgb)
plt.title("After Average Blur")
plt.axis('off')


# ----- Gaussian Blur -----
plt.subplot(3, 2, 3)
plt.imshow(img_gaussian_rgb)
plt.title("Original Image (Gaussian Blur)")
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(gaussian_blur_rgb)
plt.title("After Gaussian Blur")
plt.axis('off')


# ----- Median Blur -----
plt.subplot(3, 2, 5)
plt.imshow(img_median_rgb)
plt.title("Original Image (Median Blur)")
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(median_blur_rgb)
plt.title("After Median Blur")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np



img= cv2.imread("/content/sharpening1.jpeg",0)

# Check if image was loaded successfully
if img is None:
    print("Error: Could not load image. Please check the path and file integrity.")
else:
    kernel=np.array([[0,-1,0],
                     [-1,5,-1],
                     [0,-1,0]])

    sharpened=cv2.filter2D(img,-1,kernel)

    #Display Results

    plt.figure(figsize=(10,5))

    plt.subplot(1,3,1)
    plt.title("Original Image")
    plt.imshow(img,cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Sharpened Image")
    plt.imshow(sharpened,cmap='gray')
    plt.axis("off")
    plt.show()
```

```python
#canny and sobel image
import cv2
import matplotlib.pyplot as plt

img= cv2.imread("/content/sharpening1.jpeg",0)

# Check if image was loaded successfully
if img is None:
    print("Error: Could not load image. Please check the path and file integrity.")

#sobel edge detection
sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)#vertical edges
sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)#horizontal edges
sobel_xy=cv2.magnitude(sobel_x,sobel_y)

#canny edge detection
canny=cv2.Canny(img,100,200)

plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img,cmap='gray')
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Sobel")
plt.imshow(sobel_xy,cmap='gray')
plt.axis("off")

plt.subplot(2,2,3)
plt.title("canny")
plt.imshow(canny,cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
